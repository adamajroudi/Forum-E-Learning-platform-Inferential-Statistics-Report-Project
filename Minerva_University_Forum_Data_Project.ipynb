```python
def mean_calculation(list_variable):
    """
    Calculate the mean of a given list.

    Parameters:
    - list_variable (list): A list of numerical values.

    Returns:
    - float: The mean of the input list.
    """
    list_sum = 0  # Sum of all items in the list
    list_length = len(list_variable)  # Number of items in the list

    for i in range(list_length):
        list_sum += list_variable[i]

    return list_sum / list_length
```

```python
def median_calculation(list_variable):
    """
    Calculate the median of a given list.

    Parameters:
    - list_variable (list): A list of numerical values.

    Returns:
    - float: The median of the input list.
    """
    list_length = len(list_variable)
    list_variable.sort()

    if list_length % 2 == 0:
        # If the list length is even, calculate the average of the two middle elements
        median_right = list_variable[list_length // 2]
        median_left = list_variable[list_length // 2 - 1]
        median = (median_right + median_left) / 2
    else:
        # If the list length is odd, the median is the middle element
        median = list_variable[list_length // 2]

    return median
```

```python
def occurrences_calculation(list_variable, item):
    """
    Count the occurrences of a specific item in a list.

    Parameters:
    - list_variable (list): A list of items.
    - item: The item whose occurrences are to be counted in the list.

    Returns:
    - int: The number of occurrences of the specified item in the list.
    """
    occurrences = 0
    for i in range(len(list_variable)):
        if list_variable[i] == item:
            occurrences += 1
    return occurrences


def mode_calculation(col):
    """
    Calculate the mode(s) in a list.

    Parameters:
    - col (list): A list of items.

    Returns:
    - mixed: If there is a single mode, return the mode. If there are multiple modes, return a list of modes.
    """
    max_occurrences = 0
    mode = 0
    mode_list = []

    for i in range(len(col)):
        current_occurrences = occurrences_calculation(col, col[i])

        if current_occurrences > max_occurrences:
            max_occurrences = current_occurrences
            mode = col[i]
            mode_list = [col[i]]
        elif current_occurrences == max_occurrences and col[i] not in mode_list:
            mode_list.append(col[i])

    if len(mode_list) == 1:
        return mode
    return mode_list
```

```python
def range_calculation(list_variable):
    """
    Calculate the range of a list.

    Parameters:
    - list_variable (list): A list of numerical values.

    Returns:
    - float: The range of the input list, defined as the difference between the maximum and minimum values.
    """

    # Initialize the maximum list item with the first item of the list
    max_item = list_variable[0]

    # Initialize the minimum list item with the first item of the list
    min_item = list_variable[0]

    for i in range(len(list_variable)):
        if list_variable[i] > max_item:
            max_item = list_variable[i]
        if list_variable[i] < min_item:
            min_item = list_variable[i]

    return max_item - min_item
```

```python
def standard_deviation_calculation(list_variable):
    """
    Calculate the standard deviation of a list.

    Parameters:
    - list_variable (list): A list of numerical values.

    Returns:
    - float: The standard deviation of the input list.
    """
    S = 0  # Initialize the sum to zero

    # Calculate the mean using the mean_calculation function
    list_mean = mean_calculation(list_variable)

    for i in range(len(list_variable)):
        S += (list_variable[i] - list_mean) ** 2

    return (S / len(list_variable)) ** 0.5
```

```python
# Importing necessary functions
from tabulate import tabulate

# Subgroup 1
group1 = df[df["num_polls"] < 6]
subgroup_1 = group1['num_reactions'].values.tolist()

# Subgroup 2
group2 = df[df["num_polls"] >= 6]
subgroup_2 = group2['num_reactions'].values.tolist()

# Descriptive statistics
stats_headers = ["Statistic", "Subgroup 1", "Subgroup 2"]

statistics = [
    ["Mean", mean_calculation(subgroup_1), mean_calculation(subgroup_2)],
    ["Mode", mode_calculation(subgroup_1), mode_calculation(subgroup_2)],
    ["Median", median_calculation(subgroup_1), median_calculation(subgroup_2)],
    ["Range", range_calculation(subgroup_1), range_calculation(subgroup_2)],
    ["Standard Deviation", standard_deviation_calculation(subgroup_1), standard_deviation_calculation(subgroup_2)],
]

# Print formatted table
print(tabulate(statistics, headers=stats_headers, tablefmt="grid"))
```

```python
# importing all the necessary libraries
import pandas
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy.stats import pearsonr
import statsmodels.api as sm # useful stats package with regression functions
import seaborn as sns # plotting package

#importing the dataset
df = pandas.read_csv("https://course-resources.minerva.edu/uploaded_files/mu/00294337-9526/cornerstone-metrics-data-by-class-2021-2022--sample.csv")
df.head(0)

#data before cleaning

# Getting the descriptive statistics of instructor_talk_time_percent
prof_talktime_prec = df["instructor_talk_time_percent"]

# Getting the descriptive statistics of num_messages
num_messages =  df["num_chat_messages"]

# Get the descriptive statistics of instructor talk time percent
talktime_stats = prof_talktime_prec.describe()

# Get the median of number of instructor talk time percent
talktime_median = prof_talktime_prec.median()

# Get the descriptive statistics of number of chat messages
msgs_stats = num_messages.describe()

# Get the median of number of chat messages
msgs_median = num_messages.median()

# Print the results
print(talktime_stats, "\nmedian: ", talktime_median)
print("\n")
print(msgs_stats, "\nmedian: ", msgs_median)



# Define the regression_model function that takes two arguments: column_x and column_y
def regression_model(column_x, column_y):
    # Add a constant to the data in column_x and store it in the X variable
    X = sm.add_constant(df[column_x])
    # Store the data in column_y in the Y variable
    Y = df[column_y]
    # Fit the OLS (ordinary least squares) regression model using the data in X and Y
    regressionmodel = sm.OLS(Y,X).fit()

    # Compute and store the R-squared value, rounded to 3 decimal places
    Rsquared = round(regressionmodel.rsquared,3)
    # Compute and store the slope of the regression line, rounded to 3 decimal places
    slope = round(regressionmodel.params[1],3)
    # Compute and store the intercept of the regression line, rounded to 3 decimal places
    intercept = round(regressionmodel.params[0],3)

    # Create a subplot with two plots (scatter plot and residual plot)
    fig, (ax1, ax2) = plt.subplots(ncols=2, sharex=True, figsize=(12,4))
    # Create a scatter plot of column_x and column_y
    sns.regplot(x=column_x, y=column_y, data=df, marker="+", ax=ax1)
    # Create a residual plot of column_x and column_y
    sns.residplot(x=column_x, y=column_y, data=df, ax=ax2)
    # Set the y-label for the residual plot
    ax2.set(ylabel='Residuals')
    # Set the y-axis limits for the residual plot
    ax2.set_ylim(min(regressionmodel.resid)-1,max(regressionmodel.resid)+1)
    # Create a new figure for the histogram of the residuals
    plt.figure()
    # Create a histogram of the residuals
    sns.distplot(regressionmodel.resid, kde=False, axlabel='Residuals', color='red')

    # Print the R-squared value and the regression equation
    print("R-squared = ",Rsquared)
    print("Regression equation: "+column_y+" = ",slope,"* "+column_x+" + ",intercept)


regression_model('num_chat_messages','instructor_talk_time_percent')

#Importing the new data after
df_cleaned = pandas.read_excel("https://drive.google.com/uc?export=download&id=1-0ZosMqu8nrIm1JjGP2lbW8N0GD7KICO")

df_cleaned.head(0)


# Get the descriptive statistics of instructor_talk_time_percent
prof_talktime_prec = df_cleaned["instructor_talk_time_percent"]

# Get the descriptive statistics of num_messages
num_messages =  df_cleaned["num_chat_messages"]

#data after cleaning

# This function was used from class

def regression_model_cleaned(column_x, column_y):
    # Add a constant to the data in column_x and store it in the X variable
    X = sm.add_constant(df_cleaned[column_x])
    # Store the data in column_y in the Y variable
    Y = df_cleaned[column_y]
    # Fit the OLS (ordinary least squares) regression model using the data in X and Y
    regressionmodel = sm.OLS(Y,X).fit()

    # Compute and store the R-squared value, rounded to 3 decimal places
    Rsquared = round(regressionmodel.rsquared,3)
    # Compute and store the slope of the regression line, rounded to 3 decimal places
    slope = round(regressionmodel.params[1],3)
    # Compute and store the intercept of the regression line, rounded to 3 decimal places
    intercept = round(regressionmodel.params[0],3)

    # Create a subplot with two plots (scatter plot and residual plot)
    fig, (ax1, ax2) = plt.subplots(ncols=2, sharex=True, figsize=(12,4))
    # Create a scatter plot of column_x and column_y
    sns.regplot(x=column_x, y=column_y, data=df_cleaned, marker="+", ax=ax1)
    # Create a residual plot of column_x and column_y
    sns.residplot(x=column_x, y=column_y, data=df_cleaned, ax=ax2)
    # Set the y-label for the residual plot
    ax2.set(ylabel='Residuals')
    # Set the y-axis limits for the residual plot
    ax2.set_ylim(min(regressionmodel.resid)-1,max(regressionmodel.resid)+1)
    # Create a new figure for the histogram of the residuals
    plt.figure()
    # Create a histogram of the residuals with a KDE plot
    sns.distplot(regressionmodel.resid, kde=True, axlabel='Residuals', color='red')

    # Print the R-squared value and the regression equation
    print("R-squared = ",Rsquared)
    print("Regression equation: "+column_y+" = ",slope,"* "+column_x+" + ",intercept)

    # visualize the distribution of residuals using Q-Q plot
    sm.qqplot(regressionmodel.resid, line='s')
    plt.show()


regression_model_cleaned('num_chat_messages','instructor_talk_time_percent')

# Calculate the correlation between the two columns using Pearson's method
r = num_messages.corr(prof_talktime_prec, method='pearson')

# Print the result of Pearson's r
print("Pearson's r: ", r)

# Print the R-squared value, rounded to 3 decimal places
print("Checked R-squared", round(r**2,3))

# Calculate the sample size
n = 235

# Calculate the sample standard deviation for "num_messages" and "prof_talktime_prec"
sx = np.std(num_messages, ddof=1) # ddof=1 gives the sample standard deviation
sy = np.std(prof_talktime_prec, ddof=1)

# Calculate the slope of the regression line
b1 = round(r*(sy/sx),3)
print("b1 =",b1)

# Calculate the standard error of the slope
SE = (sy/sx)*((1-r**2)/(n-2))**0.5
print("SE =",SE)

# Calculate the critical t-value
t = stats.t.ppf(0.975,n-2) # explain why this is correct by drawing the relevant distribution
print("t =",t)

# Calculate the lower and upper bound of the 95% confidence interval for the slope
lower_bound = b1 - t*SE
upper_bound = b1 + t*SE

# Print the 95% confidence interval for the slope
print("Interval =", [round(lower_bound,3),round(upper_bound,3)])

```
